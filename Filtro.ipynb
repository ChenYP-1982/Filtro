# Para os filtros, vamos usar  sempre  a mesma condicao de BODY MASS maior que 3900 e Sex Male


```python
#fitro 1: body mass maior 3900--> traz somente valores de body mass acima de 3900
df[df["body_mass_g"] > 3900]

```

```python
#Filtrando body mass E sex-->Isso nos traz um dataframe apenas com os valores filtrados
df[(df["body_mass_g"] > 3900) & (df["sex"]=="Male")]
```

USANDO a funcao .Query()

```python
#Filtro 2:: body mass maior 3900--> traz somente valores de body mass acima de 3900
df.query("body_mass_g > 3900")
```

```python
#Filtrando body mass E sex-->Isso nos traz um dataframe apenas com os valores filtrados
df.query("body_mass_g > 3900 &  sex =='Male'")
```

Usando a funcao .LOC[]

```python
#filtro 1: body mass maior 3900--> traz somente valores de body mass acima de 3900
df.loc[df.body_mass_g > 3900]
```

```python
#Filtrando body mass E sex-->Isso nos traz um dataframe apenas com os valores filtrados
df.loc[(df.body_mass_g > 3900) & (df.sex == "Male")]
```

Para retornar o dataframe sem os valores filtrados podemos utilizar " ~ " 

```python
#retornando os valores sem os dados filtrados
df.loc[~(df.body_mass_g > 3900) & (df.sex == "Male")].head()
```
